```python
import os
from dotenv import load_dotenv
from groq import Groq

# Load environment variables
load_dotenv()

# Configure Groq
api_key = os.getenv("GROQ_API_KEY")
client = Groq(api_key=api_key)
```

```python
from rag_helper import RAGBase
from ingest import load_faq_data, build_index
```

```python
documents = load_faq_data()
index = build_index(documents)

assistant = RAGBase(index, client)
```

```python
assistant.rag("I just discovered the course, can I still join?")
```

```python
# Debug: See the search results and context being built
query = "I just discovered the course, can I still join?"
search_results = assistant.search(query)
print(f"Search results found: {len(search_results)}\n")

# Show context that's being sent to LLM
context = assistant.build_context(search_results)
print("Context sent to LLM:")
print(context)
```

```python

# Debug: See what search results are found
query = "I just discovered the course, can I still join?"
search_results = assistant.search(query)
print(f"Number of results found: {len(search_results)}\n")
print("Search Results:")
for i, result in enumerate(search_results[:3], 1):  # Show first 3
    print(f"{i}. Q: {result['question']}")
    print(f"   A: {result['answer'][:100]}...\n")
```
